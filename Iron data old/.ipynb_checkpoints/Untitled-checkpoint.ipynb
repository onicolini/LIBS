```python
from glob import glob
filenames = glob('./data/0*')
dfs = [pd.read_csv(f,sep='\t') for f in filenames]
#type(dfs) #dfs is a list of dataframes 
```

```python
df=dfs[0].drop(columns=['Pixels'])
df=df.dropna(axis=1)
data=df.values
plt.plot(data[0], data[400])
```

```python
ydf=df.drop([0, 1])
yval=ydf.values
yval
```

```python
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(yval)
X_std
```

```python
mean_vec = np.mean(X_std, axis=0)
type(mean_vec)
len(mean_vec)
```

```python
cov_mat = (X_std - mean_vec).T.dot((X_std - mean_vec)) / (X_std.shape[0]-1)
print('Covariance matrix \n%s' %cov_mat)
```

```python
cov_mat = np.cov(X_std.T)
#print('NumPy covariance matrix: \n%s' %np.cov(X_std.T))

eig_vals, eig_vecs = np.linalg.eig(cov_mat)

print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
u,s,v = np.linalg.svd(X_std.T)
u
```

```python
for ev in eig_vecs.T:
    np.testing.assert_array_almost_equal(1.0, np.linalg.norm(ev))
print('Everything ok!')
```

```python
# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort(key=lambda x: x[0], reverse=True)

# Visually confirm that the list is correctly sorted by decreasing eigenvalues
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
```

```python
tot = sum(eig_vals)
#print(tot)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
```

```python
from matplotlib import pyplot as plt
import math

with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(6, 4))

    plt.bar(range(4095), var_exp, alpha=0.5, align='center', label='individual explained variance')
    plt.step(range(4095), cum_var_exp, where='mid', label='cumulative explained variance')
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc='best')
    plt.tight_layout()
```
